```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
upd_df = pd.read_csv('ndtv_data_final.csv')

# Selecting the desired columns
selected_columns = ['Operating system', 'Price', 'Front camera', 'Wi-Fi']
selected_df = upd_df[selected_columns]

# Encode categorical variables (one-hot encoding)
selected_df_encoded = pd.get_dummies(selected_df)

# Create a heatmap of the correlation matrix with a different color palette
plt.figure(figsize=(10, 8))
sns.heatmap(data=selected_df_encoded.corr(), annot=True, cmap='viridis', fmt=".2f", vmin=-1, vmax=1 ,linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = upd_df.select_dtypes(include=['int', 'float']).columns
selected_df = upd_df[numeric_columns]

plt.figure(figsize=(10, 6))
sns.heatmap(data=selected_df.corr(), annot=True, fmt=".2f", vmin=-1, vmax=1 ,linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
